# Convert PDF to an Image

To better work with the contract, let's try to convert the pdf to an image.


```python
import pdf2image  # For pdf2image, make sure poppler is installed

base_path = "./data/"

images = pdf2image.convert_from_path(f"{base_path}contract1.pdf")
for i, image in enumerate(images):
    image.save(f"{base_path}contract1_{i}.jpg", "JPEG")
images[1]
```
